


# **Local Host Security Scan**
## **Project Overview**
This project is a quick **Nmap scan** of the **localhost (127.0.0.1)** to identify open ports and active services running on the machine.  
It simulates a **real-world use case** where a cybersecurity practitioner periodically checks their system for unexpected services or suspicious activity.

---

## **Why Perform a Local Host Security Scan?**
1. **Routine Security Check** – Ensure only necessary services are running and no unexpected open ports exist.  
2. **Troubleshooting & Diagnostics** – Verify if required services (e.g., SSH, RDP, web servers) are active.  
3. **Incident Response** – Investigate potential **malware activity** or unauthorized access if something seems unusual (e.g., sluggish system, strange connections).

Key Findings & Security Considerations
SSH (Port 22) is Open

Used for Secure Shell (SSH).
If not in use, consider disabling SSH.
Microsoft RPC (135) and SMB (445) are Open

Could be a security risk if SMB is enabled but not needed.
Intel AMT (16992) is Open

Remote management is enabled, check BIOS settings.
High-numbered dynamic RPC ports (49664-49670) are open

Likely Windows services, but should be monitored for unusual activity.

Next Steps
Disable unnecessary services to reduce attack surface.
Configure Windows Firewall rules to block unused ports.
Check BIOS settings for Intel AMT and disable if not needed.
